```python
!pip install pandas
import pandas as pd
tabela_Harry=pd.read_csv('HarryPotter.csv')
```

```python
display(tabela_Harry.info())
```

```python
tabela_Harry=tabela_Harry.drop('image', axis =1)
tabela_Harry=tabela_Harry.drop('dateOfBirth', axis =1)
```

```python
display(tabela_Harry.info())
```

```python
tabela_Harry['house'].value_counts()
```

```python
tabela_Harry = tabela_Harry.dropna(how='any', axis= 0 )
```

```python
display(tabela_Harry.info())
```

```python
tabela_Harry['patronus'].value_counts()
```

```python
!pip install plotly
import plotly.express as px 
```

```python
grafico=px.histogram(tabela_Harry, x='name' , color='patronus',  title='Histograma Harry Potter', opacity=0.5,)
grafico.show()
```

```python
for coluna in tabela_Harry.columns:  
    grafico =px.histogram(tabela_Harry, x=coluna ,title='Histograma Harry Potter', opacity=0.5, color='name') 
    grafico.show()
```
