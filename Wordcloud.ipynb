 ## Importação e preparação 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
dados= pd.read_csv('palavras.csv')
dados= pd.DataFrame(dados)
dados.head
```

```python
words = dados['palavras']
all_words = " ".join(w for w in words)
```

```python
stopwords = set(STOPWORDS)
```

## definindo uma função que irá plotar a wordcloud

```python

def plot_wordcloud(wc):
  fig = plt.figure(dpi=300)
  fig, ax = plt.subplots(figsize=(12,6))
  ax.imshow(wc, interpolation = 'bilinear')
  ax.set_axis_off()
  plt.imshow(wc)
```

## Customizando cores 01 - passando diretamento como parâmetro


```python

wc  = WordCloud(stopwords=stopwords, background_color='black', 
             width=1600, height=800, colormap= 'PuBuGn', max_font_size=60, min_font_size=20). generate(all_words)
imagem = plot_wordcloud(wc)
```

## Customizando cores 02 - a partir de uma lista de cores

```python
colors= [ 'navy', 'purple', 'green', 'pink','gray', 'white', 'orange']
my_cmap = ListedColormap(sns.color_palette(colors).as_hex())
```

```python
wc=WordCloud(stopwords=stopwords, background_color='black', colormap = my_cmap,
             width=1600, height=800, max_font_size=60, min_font_size=10). generate(all_words)
plot_wordcloud(wc)
```
